## Check Null Values in Dataset

```python
df.isnull().sum()
```

## Create 2 Columns

```python
 df['MinsPerMatch'] = df['Mins'] / df['Matches'].astype(int)
df['GoalsPerMatch'] = df['Goals'] / df['Matches'].astype(float)
```

```python
df.head()
```

## Count Insights Data per Competetion

```python
# ==> Count Goals
total_goals = df['Goals'].sum()
total_goals
```

```python
# ==> Count Goals from Penalty

total_goals_penalty = df['Penalty_Goals'].sum()
total_goals_penalty
```

```python
# ==> Count total penality Attemps

total_penality_attemps = df['Penalty_Attempted'].sum()
total_penality_attemps
```

## Visulization of penality Scored | Missed

```python
plt.figure(figsize=(13,6))
dt = df['Penalty_Attempted'].sum()-total_goals_penalty
data = [dt,total_goals_penalty]
label = ['Penalities Missed', 'Penalities Scored']
color = sns.color_palette('Set2')
plt.pie(data,labels=label,colors=color,autopct="%.0f%%")
```
